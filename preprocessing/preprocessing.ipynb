```python
import altair as alt
import pandas as pd
```

```python
data = pd.read_csv("data/delays_2019.csv")
```

```python
subset = data.loc[:,["carrier_name","arr_flights","arr_cancelled", "arr_diverted","date"]]
subset[['year', 'month']] = subset['date'].str.split('-', 1, expand=True)
subset = subset.drop(["date","year"], axis=1).groupby(['carrier_name',"month"], as_index=False)[["arr_flights","arr_cancelled","arr_diverted"]].apply(sum)
```

```python
subset["month"] = pd.to_numeric(subset["month"])
subset["percent_cancelled"] = round(subset["arr_cancelled"]/subset["arr_flights"]*100,2)
subset["percent_diverted"] = round(subset["arr_diverted"]/subset["arr_flights"]*100,2)
```

```python
subset
```

```python
subset.to_csv("Preprocessed_data.csv")
```

```python
plot = subset.loc[subset["carrier_name"]=="Alaska Airlines Inc.",:]
alt.Chart(plot, title='Alaska Airlines - 2019').mark_line(size=3, color = "red").encode(
     alt.X("month", title = "Month"),
     alt.Y("percent_cancelled", title = "Percent of Flights Cancelled")
    
)
```

```python
def plot_altair(carrier_name):
    plot = data.loc[data["carrier_name"].isin(carrier_name),:]
    chart = alt.Chart(plot, title= "Cancelled Flights -2019").mark_line(size=3).encode(
         alt.X("month", title = "Month"),
         alt.Y("percent_cancelled", title = "Percent of Flights Cancelled"),
         color = alt.Color("carrier_name", title = "Carrier Name")
    )
    return chart
```
